```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

Load both train and test sets

```python
train_df = pd.read_csv("train_df.csv")
test_df = pd.read_csv("test_df.csv")
```

Drop columns with no distinct values

```python
columns_to_drop = ["feature_0", "feature_73", "feature_74", "feature_75"]
train_df.drop(columns=columns_to_drop, inplace=True)
test_df.drop(columns=columns_to_drop, inplace=True)
```

```python
train_df
```

```python
test_df
```

Scale the data to get rid of different scales of different features.
MinMaxScaler is chosen because features' distributions are far from being normal that's why we want to keep them and keeping values between 0 and 1.

```python
# get list of features to scale having list of features to exclude
features_to_exclude = ["search_id", "target"]
features_to_scale = [col for col in train_df.columns if col not in features_to_exclude]
```

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler is being fitted only only train data
scaler = MinMaxScaler()
scaler.fit(train_df[features_to_scale])

# scale both train and test sets
train_df[features_to_scale] = scaler.transform(train_df[features_to_scale])
test_df[features_to_scale] = scaler.transform(test_df[features_to_scale])
```

```python
train_df.head()
```

```python
test_df.head()
```

Check for linear dependence

```python
# get correlation matrix of train set
corr_matrix = train_df[features_to_scale].corr()

# plot heatmap with annotations
plt.figure(figsize=(30, 30))
sns.heatmap(corr_matrix, annot=True, square=True, fmt=".2f", cbar_kws={"shrink": 0.5})
plt.show()
```

Get the number of highly correlated pairs of features.
Threshold is taken as 0.8 level as some standart value.

```python
# default treshold for detecting high correlation
threshold = 0.8

# check for direct correlations amount all values excluding the diagonal
direct_correlations = (
    len(np.where(corr_matrix >= threshold)[0])
    - len(train_df[features_to_scale].columns)
) // 2

# check for inverse correlations
inverse_correlations = (len(np.where(corr_matrix <= -threshold)[0])) // 2

print(direct_correlations, inverse_correlations)
```

```python
# get indices of values in correlation matrix that are greater than threshold
temp = np.where((corr_matrix >= threshold) & (corr_matrix != 1.0))

# rotate to get pairs of indices
temp = np.stack(temp, axis=1)
# temp = temp[1]

# take only first half, don't repeat pairs with reversed indices
temp = temp[: len(temp) // 2 - 1]

temp
```

As we see some features have multiple linear dependencies

```python
features_to_drop_indices = temp.reshape((1, -1)).ravel()

features_to_drop_indices, features_to_drop_indices.shape
```

```python
feature_names_to_drop = corr_matrix.columns[features_to_drop_indices]

feature_names_to_drop, feature_names_to_drop.shape
```

Simply drop these highly correlated features

```python
train_df.drop(columns=feature_names_to_drop, inplace=True)
test_df.drop(columns=feature_names_to_drop, inplace=True)
```

According to distribution here is the proposal to encode firstly search_ids where target == 1 exists at least ones.

The logic: if this object was selected at least ones it is more probably that is has something important to consider

```python
temp = train_df[["search_id", "target"]]

# ! a crutch to extract the needed ids
temp = temp.groupby("search_id").filter(lambda x: (x["target"] == 1).any())
temp[temp["target"] == 1]
temp.drop_duplicates(inplace=True)

target_equals_1_at_least_ones_search_ids = list(set(temp["search_id"]))
# len(target_equals_1_at_least_ones_search_ids) # 213
```

```python
target_equals_1_at_least_ones_search_ids[:5]
```

One Hot Encode these at least ones appeared search ids

```python
# filter to all the rows where search id at least ones has a value 1 in target
filtered_train_df = train_df[
    train_df["search_id"].isin(target_equals_1_at_least_ones_search_ids)
]

# one hot encode itself
encoded_train_df = pd.get_dummies(
    filtered_train_df["search_id"], prefix="search_id", dtype=np.int8
)

# concatenate the original with encoded
encoded_train_df = pd.concat([train_df, encoded_train_df], axis=1)

# fill NaN values with zeros where for search_id there are no target = 1
encoded_train_df.fillna(0, inplace=True)

# drop categorical feature search_id
encoded_train_df.drop(["search_id"], axis=1, inplace=True)
```

```python
# filter to all the rows where search id at least ones has a value 1 in target
filtered_test_df = test_df[
    test_df["search_id"].isin(target_equals_1_at_least_ones_search_ids)
]

# one hot encode itself
encoded_test_df = pd.get_dummies(
    filtered_test_df["search_id"], prefix="search_id", dtype=np.int8
)

# concatenate the original with encoded
encoded_test_df = pd.concat([test_df, encoded_test_df], axis=1)

# fill NaN values with zeros where for search_id there are no target = 1
encoded_test_df.fillna(0, inplace=True)

# drop categorical feature search_id
encoded_test_df.drop(["search_id"], axis=1, inplace=True)
```
